```python
bios_new['born_year']=bios_new['born_datetime'].dt.year #extracting year from born_datetime
bios_new.head()
```

```python
bios_new['height_category']=bios_new['height_cm'].apply(lambda x: 'Short' if x<165 else('Average' if x<175 else 'Tall'))
bios_new.head()
```

```python
def categorize_athlete(row):
  if row['height_cm'] <175 and row['weight_kg'] <70:
    return 'LightBuilt'
  elif row['height_cm']<185 and row['weight_kg']<=80:
    return 'MediumBuilt'
  else:
    return 'HeavyBuilt'
bios_new['Category']=bios_new.apply(categorize_athlete,axis=1) #axis 0 is column 1 is row
bios_new.head(5)
```

```python
bios_new.to_csv('/content/sample_data/bios_modified.csv',index=False)
```
